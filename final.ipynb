```python
sorted_data = data.sort_values(by=['Date', 'instrument_type', 'strike', 'time']).reset_index(drop=True)
sorted_data.head(60)
```

```python
import pandas as pd
import numpy as np

data = sorted_data.copy()  # Made a copy to avoid modifying original data
data['time'] = pd.to_datetime(data['time'])

time_range = pd.date_range(start='09:15:00', end='15:30:00', freq='5T')
expanded_data = pd.DataFrame()

grouped_data = data.groupby(['Date', 'instrument_type'])


for (Date, instrument_type), subset in grouped_data:
    # Group the subset by strike and resample each group to 5-minute intervals
    resampled_subset = subset.set_index('time').groupby('strike').resample('5T').agg({
        'open': 'first',
        'high': 'max',
        'low': 'min',
        'close': 'last',
        'volume': 'sum',
        'oi': 'sum'
    }).reset_index(level='strike')

    # Forward filling
    resampled_subset[['high', 'low']] = resampled_subset[['high', 'low']].ffill()
    # Interpolating
    resampled_subset[['volume', 'oi']] = resampled_subset[['volume', 'oi']].replace(0, np.nan)
    resampled_subset[['volume', 'oi']] = resampled_subset[['volume', 'oi']].interpolate(method='linear')
    resampled_subset[['volume', 'oi']] = resampled_subset[['volume', 'oi']].fillna(0).round(2)
    resampled_subset[['open', 'close']] = resampled_subset[['open', 'close']].interpolate(method='linear').round(2)

    resampled_subset.reset_index(inplace=True)

    for col in ['symbol', 'expiry']:
        resampled_subset[col] = subset[col].iloc[0]

    resampled_subset['Date'] = Date
    resampled_subset['instrument_type'] = instrument_type
    expanded_data = pd.concat([expanded_data, resampled_subset], ignore_index=True)

expanded_data.sort_values(by=['Date', 'instrument_type', 'strike', 'time'], inplace=True)
expanded_data.reset_index(drop=True, inplace=True)
expanded_data['time'] = pd.to_datetime(expanded_data['time']).dt.time
```

```python
import pandas as pd
def preprocess_date(date_string):
    date_str = str(date_string)
    if len(date_str) == 8:
        # If the date string is in the format 'DDMMYYYY', convert it to 'DD-MM-YYYY'
        return date_str[:2] + '-' + date_str[2:4] + '-' + date_str[4:]
    else:
        date_str = '0' + date_str
        return date_str[:2] + '-' + date_str[2:4] + '-' + date_str[4:]
expanded_data['Preprocessed_Date'] = expanded_data['Date'].apply(preprocess_date)
```

```python
import pandas as pd
import os
from datetime import datetime

directory_path = '/content/gdrive/MyDrive/Fut_data'
data_list = []

for filename in os.listdir(directory_path):
    if filename.endswith('.csv'):
        try:
            date_str = filename.split('.')[0]  # Assuming the filename is like 'DDMMYYYY.csv'
            date_obj = datetime.strptime(date_str, '%d%m%Y')
            formatted_date = date_obj.strftime('%d-%m-%Y')
        except ValueError:
            print(f'{filename}: Filename does not match expected date format')
            continue

        if date_obj.year==2018 and date_obj.month==12: # Considering only a part of the data
            file_path = os.path.join(directory_path, filename)
            df = pd.read_csv(file_path)

            # Considering ATM value from the first row
            if 'open' in df.columns:
                atm_value = df['open'].iloc[0]
                # Appending the formatted date and the ATM value to the data list
                data_list.append((formatted_date, atm_value))
            else:
                print(f'{formatted_date}: "open" column not found')

data_list.sort(key=lambda x: datetime.strptime(x[0], '%d-%m-%Y'))
for formatted_date, atm_value in data_list:
    print(f'{formatted_date}: ATM = {atm_value}')
```

```python
import pandas as pd
def calculate_pcr(df, atm_strike):
    lower_bound = atm_strike - 1000
    upper_bound = atm_strike + 1000
    selected_strikes = df[(df['strike'] >= lower_bound) & (df['strike'] <= upper_bound)]

    # Filter data for entire day
    selected_day = selected_strikes[selected_strikes['Preprocessed_Date'] == selected_strikes['Preprocessed_Date'].iloc[0]]

    total_put_volume = selected_day[selected_day['instrument_type'] == 'PE']['volume'].sum()
    total_call_volume = selected_day[selected_day['instrument_type'] == 'CE']['volume'].sum()
    pcr = total_call_volume / total_put_volume if total_put_volume != 0 else None
    return pcr

def get_unique_combinations(df):
    return df[['Preprocessed_Date', 'time']].drop_duplicates()

# Iterating over unique combinations and calculate PCR
for _, combo in get_unique_combinations(df).iterrows():
    preprocessed_date = combo['Preprocessed_Date']
    time = combo['time']
    atm_strike = df[(df['Preprocessed_Date'] == preprocessed_date) & (df['time'] == time)]['ATM'].iloc[0]
    pcr = calculate_pcr(df, atm_strike)
    print(f"Preprocessed_Date: {preprocessed_date}, Time: {time}, PCR: {pcr}")
```
