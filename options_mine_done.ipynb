```python
import glob
import csv
import os
import time

start = time.time()

path = "/content/gdrive/MyDrive/bank_nifty data/csv"
output_file = 'dataset.csv'
cols = ['time', 'symbol', 'open', 'high', 'low', 'close', 'volume', 'oi', 'expiry', 'strike', 'instrument_type', 'Date']

# List of file endings to consider
file_endings = ['012018', '022018', '032018']

with open(output_file, 'w', newline='') as out:
    writer = csv.writer(out)
    writer.writerow(cols)

# Iterate through files with specified endings
for ending in file_endings:
    all_files = glob.glob(os.path.join(path, f"*{ending}.csv"))
    for file in all_files:
        date = os.path.basename(file).split('.')[0]
        with open(file, 'r') as f, open(output_file, 'a', newline='') as out:
            reader = csv.reader(f)
            writer = csv.writer(out)
            next(reader)  # Skip the header row
            for row in reader:
                row.append(date)
                writer.writerow(row)

end = time.time()
print("Execution time:", end - start)
```

```python
import pandas as pd
def preprocess_date(date_string):
    date_str = str(date_string)
    if len(date_str) == 8:
        # If the date string is in the format 'DDMMYYYY', convert it to 'DD-MM-YYYY'
        return date_str[:2] + '-' + date_str[2:4] + '-' + date_str[4:]
    else:
        date_str = '0' + date_str
        return date_str[:2] + '-' + date_str[2:4] + '-' + date_str[4:]
data['Preprocessed_Date'] = data['Date'].apply(preprocess_date)
```

```python
sorted_data = data.sort_values(by=['Preprocessed_Date', 'instrument_type', 'strike', 'time']).reset_index(drop=True)
sorted_data.head(60)
```

```python
import pandas as pd
import numpy as np

data = sorted_data.copy()  # Made a copy to avoid modifying original data
data['time'] = pd.to_datetime(data['time'])

time_range = pd.date_range(start='09:15:00', end='15:30:00', freq='5T')
expanded_data = pd.DataFrame()

grouped_data = data.groupby(['Preprocessed_Date', 'instrument_type'])


for (Preprocessed_Date, instrument_type), subset in grouped_data:
    # Group the subset by strike and resample each group to 5-minute intervals
    resampled_subset = subset.set_index('time').groupby('strike').resample('5T').agg({
        'open': 'first',
        'high': 'max',
        'low': 'min',
        'close': 'last',
        'volume': 'sum',
        'oi': 'last'
    }).reset_index(level='strike')

    # Forward filling
    resampled_subset[['high', 'low']] = resampled_subset[['high', 'low']].ffill()
    # Interpolating
    resampled_subset[['volume', 'oi']] = resampled_subset[['volume', 'oi']].replace(0, np.nan)
    resampled_subset[['volume', 'oi']] = resampled_subset[['volume', 'oi']].interpolate(method='linear')
    resampled_subset[['volume', 'oi']] = resampled_subset[['volume', 'oi']].fillna(0).round(2)
    resampled_subset[['open', 'close']] = resampled_subset[['open', 'close']].interpolate(method='linear').round(2)

    resampled_subset.reset_index(inplace=True)

    for col in ['symbol', 'expiry']:
        resampled_subset[col] = subset[col].iloc[0]

    resampled_subset['Preprocessed_Date'] = Preprocessed_Date
    resampled_subset['instrument_type'] = instrument_type
    expanded_data = pd.concat([expanded_data, resampled_subset], ignore_index=True)

expanded_data.sort_values(by=['Preprocessed_Date', 'instrument_type', 'strike', 'time'], inplace=True)
expanded_data.reset_index(drop=True, inplace=True)
expanded_data['time'] = pd.to_datetime(expanded_data['time']).dt.time
```

```python
import pandas as pd

def calculate_intraday_pcr(df):
    # Ensure 'Preprocessed_Date' and 'time' are strings
    df['Preprocessed_Date'] = df['Preprocessed_Date'].astype(str)
    df['time'] = df['time'].astype(str)

    # Combine 'Preprocessed_Date' and 'time' to ensure 'datetime' is in datetime format
    df['datetime'] = pd.to_datetime(df['Preprocessed_Date'] + ' ' + df['time'], format='%d-%m-%Y %H:%M:%S')

    # Get the unique 5-minute intervals in the data
    time_intervals = df['datetime'].dt.floor('5T').unique()

    for time in time_intervals:
        # Filter data for the current 5-minute interval
        interval_df = df[df['datetime'].dt.floor('5T') == time]

        atm_strike = interval_df['atm'].iloc[0] if not interval_df['atm'].empty else None

        if atm_strike is not None:
            lower_bound = atm_strike - 1000
            upper_bound = atm_strike + 1000
            selected_strikes = interval_df[(interval_df['strike'] >= lower_bound) & (interval_df['strike'] <= upper_bound)]

            total_put_volume = selected_strikes[selected_strikes['instrument_type'] == 'PE']['volume'].sum()
            total_call_volume = selected_strikes[selected_strikes['instrument_type'] == 'CE']['volume'].sum()

            pcr = total_put_volume / total_call_volume if total_call_volume != 0 else None

            # Update the 'intraday_pcr' for the filtered interval data
            df.loc[df['datetime'].dt.floor('5T') == time, 'intraday_pcr'] = pcr

    return df

# Assuming expanded_data is your initial dataset
merged_df = pd.DataFrame(expanded_data)

# Calculate intraday PCR and update the DataFrame
merged_df = calculate_intraday_pcr(expanded_df)

merged_df
```

```python
import pandas as pd

def calculate_cumulative_pcr(df):
    # Combine 'Preprocessed_Date' and 'time' to ensure 'time' is in datetime format
    df['datetime'] = pd.to_datetime(df['Preprocessed_Date'] + ' ' + df['time'], format='%d-%m-%Y %H:%M:%S')

    # Initialize cumulative put and call volumes
    cumulative_put_volume = 0
    cumulative_call_volume = 0

    # Get the unique 5-minute intervals in the data
    time_intervals = df['datetime'].dt.floor('5T').unique()

    for time in time_intervals:
        # Filter data for the current 5-minute interval
        interval_df = df[df['datetime'].dt.floor('5T') == time]

        atm_strike = interval_df['atm'].iloc[0] if not interval_df['atm'].empty else None

        if atm_strike is not None:
            lower_bound = atm_strike - 1000
            upper_bound = atm_strike + 1000
            selected_strikes = interval_df[(interval_df['strike'] >= lower_bound) & (interval_df['strike'] <= upper_bound)]

            interval_put_volume = selected_strikes[selected_strikes['instrument_type'] == 'PE']['volume'].sum()
            interval_call_volume = selected_strikes[selected_strikes['instrument_type'] == 'CE']['volume'].sum()

            # Update cumulative volumes
            cumulative_put_volume += interval_put_volume
            cumulative_call_volume += interval_call_volume

            # Calculate cumulative PCR
            cumulative_pcr = cumulative_put_volume / cumulative_call_volume if cumulative_call_volume != 0 else None

            # Update the 'Cumulative_pcr' for the filtered interval data
            df.loc[df['datetime'].dt.floor('5T') == time, 'cumulative_pcr'] = cumulative_pcr

    return df

# Assuming merged_df is your DataFrame after merging the most active strikes
final_merged_df = pd.DataFrame(merged_df)

# Calculate cumulative PCR and update the DataFrame
final_merged_df = calculate_cumulative_pcr(final_merged_df)
final_merged_df
```
