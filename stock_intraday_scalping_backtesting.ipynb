```python
import pandas as pd
from fyers_apiv3 import fyersModel
import time

client_id = "6HK249XFBQ-100"
access_token = "eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJhdWQiOlsiZDoxIiwiZDoyIiwieDowIiwieDoxIiwieDoyIl0sImF0X2hhc2giOiJnQUFBQUFCb1lZbjZMSTNSanljNFlNRlhMSkZVMnN0N21VNnhVYXlCWVhsUE9HMVJVWldIUXhVVV82OFZmV2ctUEFtRFd2N1RxeFY3OF95dHY1ZXlIQjMzcXM4WnduMnRNQjdHWFhULWVmV3pBaGM5M252S0VZZz0iLCJkaXNwbGF5X25hbWUiOiIiLCJvbXMiOiJLMSIsImhzbV9rZXkiOiIxNGUwYzFlNGYzZTU0MDUzNGI4YzJkMDA1YjdlMTU0ODg0M2NhMzJkYmFhZjc2ZjQzMDQ5YzdlMCIsImlzRGRwaUVuYWJsZWQiOiJOIiwiaXNNdGZFbmFibGVkIjoiTiIsImZ5X2lkIjoiRkFBODM3MDUiLCJhcHBUeXBlIjoxMDAsImV4cCI6MTc1MTMyOTgwMCwiaWF0IjoxNzUxMjIyNzc4LCJpc3MiOiJhcGkuZnllcnMuaW4iLCJuYmYiOjE3NTEyMjI3NzgsInN1YiI6ImFjY2Vzc190b2tlbiJ9.t0TyPqSW3TrobYZYrY9vwQF1otcHbL_YocNg2J7JTeE"

# Load filtered symbols from previous backtest result

#symbols_df = pd.read_csv("backtested_trades_new1.csv")
#fyers_symbols = symbols_df["symbol"].unique().tolist()

fyers_symbols=fyers_symbols_filtered

# Initialize Fyers API
fyers = fyersModel.FyersModel(client_id=client_id, token=access_token)

# Store all trades
all_trades = []

for symbol in fyers_symbols:
    print(f"🔍 Checking: {symbol}")
    try:
        # Fetch OHLCV data
        payload = {
            "symbol": symbol,
            "resolution": "5",
            "date_format": "1",
            "range_from": "2025-06-30",
            "range_to": "2025-06-30",
            "cont_flag": "1"
        }
        data = fyers.history(payload)
        if data["s"] != "ok":
            print(f"❌ Failed to fetch {symbol}: {data}")
            continue

        df = pd.DataFrame(data["candles"], columns=["timestamp", "open", "high", "low", "close", "volume"])
        df["time"] = pd.to_datetime(df["timestamp"], unit="s").dt.tz_localize("UTC").dt.tz_convert("Asia/Kolkata")
        df.set_index("time", inplace=True)

        trade_entered = False

        for i in range(len(df) - 2):  # leave space for entry & trailing
            if trade_entered:
                break

            row = df.iloc[i]
            time_only = row.name.time()
            if not (pd.to_datetime("10:00").time() <= time_only <= pd.to_datetime("14:30").time()):
                continue

            open_, close, volume = row["open"], row["close"], row["volume"]
            move_pct = (close - open_) / open_

            # Turnover condition
            turnover = close * volume
            if turnover < 1e7:  # 1 Cr = 10^7
                continue

            if move_pct >= 0.01:
                entry_candle = df.iloc[i + 1]
                entry_price = entry_candle["open"]

                # Set SL (dynamic)
                sl = entry_price - 0.01 * entry_price if move_pct > 0.02 else open_

                # Trail SL as price goes up
                for j in range(i + 2, len(df)):
                    next_candle = df.iloc[j]
                    if next_candle["low"] <= sl:
                        exit_price = sl
                        pnl = round((exit_price - entry_price) / entry_price * 100, 2)
                        all_trades.append({
                            "symbol": symbol,
                            "trigger_time": row.name,
                            "entry_time": entry_candle.name,
                            "entry_price": entry_price,
                            "exit_time": next_candle.name,
                            "exit_price": exit_price,
                            "pnl_%": pnl
                        })
                        trade_entered = True
                        break
                    elif next_candle["close"] > entry_price:
                        # Update SL to 1% below next candle close
                        sl = max(sl, next_candle["close"] - 0.01 * next_candle["close"])

                else:
                    # If no SL hit till end, exit on last candle close
                    final_exit = df.iloc[-1]
                    exit_price = final_exit["close"]
                    pnl = round((exit_price - entry_price) / entry_price * 100, 2)
                    all_trades.append({
                        "symbol": symbol,
                        "trigger_time": row.name,
                        "entry_time": entry_candle.name,
                        "entry_price": entry_price,
                        "exit_time": final_exit.name,
                        "exit_price": exit_price,
                        "pnl_%": pnl
                    })
                    trade_entered = True

        time.sleep(3.5)  # Respect rate limit

    except Exception as e:
        print(f"⚠️ Error in {symbol}: {e}")
        continue

# Final output
trades_df = pd.DataFrame(all_trades)
#trades_df.to_csv("bullish_trailing_strategy_results.csv", index=False)
print("\n✅ Strategy Complete. Total trades:", len(trades_df))
print(trades_df)
```
