```python
# Calculate VWAP support and resistance rejections with PCR ranges and separate them based on success and failure
def calculate_vwap_rejection_pcr_ranges(df):
    support_pcrs = {'successful': [], 'failed': []}
    resistance_pcrs = {'successful': [], 'failed': []}
    support_counts = {'successful': 0, 'failed': 0}
    resistance_counts = {'successful': 0, 'failed': 0}
    ce_oi_support = {'successful': [], 'failed': []}
    pe_oi_resistance = {'successful': [], 'failed': []}

    df['prev_day_vwap'] = df['vwap'].shift(1)
    df.dropna(subset=['prev_day_vwap'], inplace=True)

    for index in range(1, len(df)):
        previous_close = df.iloc[index - 1]['close']
        current_close = df.iloc[index]['close']
        prev_day_vwap = df.iloc[index]['prev_day_vwap']
        intraday_pcr = df.iloc[index]['intraday_pcr']
        ce_oi = df.iloc[index]['ce_oi']
        pe_oi = df.iloc[index]['pe_oi']
        current_time = df.iloc[index]['datetime'].time()

        # Check if the candle is within 0.5% of the previous day's VWAP
        within_range = abs((current_close - prev_day_vwap) / prev_day_vwap) <= 0.005

        # Price takes VWAP as support
        if within_range and current_close > previous_close:
            end_time = pd.Timestamp.combine(df.iloc[index]['datetime'].date(), pd.Timestamp('15:30:00').time())
            subsequent_high = df.iloc[index + 1:index + 16]
            if (subsequent_high['datetime'] <= end_time).all():
                subsequent_high = subsequent_high['high'].max()
            else:
                subsequent_high = df[(df['datetime'] > df.iloc[index]['datetime']) &
                                     (df['datetime'] <= end_time)]['high'].max()

            if (subsequent_high - current_close) / current_close >= 0.003:
                support_pcrs['successful'].append(intraday_pcr)
                support_counts['successful'] += 1
                ce_oi_support['successful'].append(ce_oi)
            elif (subsequent_high - prev_day_vwap) / prev_day_vwap >= 0.002:
                support_pcrs['failed'].append(intraday_pcr)
                support_counts['failed'] += 1
                ce_oi_support['failed'].append(ce_oi)

        # Price takes VWAP as resistance
        if within_range and current_close < previous_close:
            end_time = pd.Timestamp.combine(df.iloc[index]['datetime'].date(), pd.Timestamp('15:30:00').time())
            subsequent_low = df.iloc[index + 1:index + 16]
            if (subsequent_low['datetime'] <= end_time).all():
                subsequent_low = subsequent_low['low'].min()
            else:
                subsequent_low = df[(df['datetime'] > df.iloc[index]['datetime']) &
                                    (df['datetime'] <= end_time)]['low'].min()

            if (current_close - subsequent_low) / current_close >= 0.003:
                resistance_pcrs['successful'].append(intraday_pcr)
                resistance_counts['successful'] += 1
                pe_oi_resistance['successful'].append(pe_oi)
            elif (prev_day_vwap - subsequent_low) / prev_day_vwap >= 0.002:
                resistance_pcrs['failed'].append(intraday_pcr)
                resistance_counts['failed'] += 1
                pe_oi_resistance['failed'].append(pe_oi)

    return {
        'support_pcrs': support_pcrs,
        'resistance_pcrs': resistance_pcrs,
        'support_counts': support_counts,
        'resistance_counts': resistance_counts,
        'ce_oi_support': ce_oi_support,
        'pe_oi_resistance': pe_oi_resistance
    }

# Group by date and calculate VWAP rejections and failures with PCR ranges for each group
pcr_results = data.groupby(data['datetime'].dt.date).apply(calculate_vwap_rejection_pcr_ranges).reset_index()
pcr_results.columns = ['date', 'pcr_ranges']

# Initialize accumulators for the results
support_pcrs = {'successful': [], 'failed': []}
resistance_pcrs = {'successful': [], 'failed': []}
support_counts = {'successful': 0, 'failed': 0}
resistance_counts = {'successful': 0, 'failed': 0}
ce_oi_support = {'successful': [], 'failed': []}
pe_oi_resistance = {'successful': [], 'failed': []}

# Combine results from all groups
for result in pcr_results['pcr_ranges']:
    for key in support_pcrs:
        support_pcrs[key].extend(result['support_pcrs'][key])
        resistance_pcrs[key].extend(result['resistance_pcrs'][key])
    for key in support_counts:
        support_counts[key] += result['support_counts'][key]
        resistance_counts[key] += result['resistance_counts'][key]
    for key in ce_oi_support:
        ce_oi_support[key].extend(result['ce_oi_support'][key])
        pe_oi_resistance[key].extend(result['pe_oi_resistance'][key])

# Calculate averages
average_support_successful_pcr = sum(support_pcrs['successful']) / len(support_pcrs['successful']) if support_pcrs['successful'] else 0
average_support_failed_pcr = sum(support_pcrs['failed']) / len(support_pcrs['failed']) if support_pcrs['failed'] else 0
average_resistance_successful_pcr = sum(resistance_pcrs['successful']) / len(resistance_pcrs['successful']) if resistance_pcrs['successful'] else 0
average_resistance_failed_pcr = sum(resistance_pcrs['failed']) / len(resistance_pcrs['failed']) if resistance_pcrs['failed'] else 0

average_ce_oi_support_successful = sum(ce_oi_support['successful']) / len(ce_oi_support['successful']) if ce_oi_support['successful'] else 0
average_ce_oi_support_failed = sum(ce_oi_support['failed']) / len(ce_oi_support['failed']) if ce_oi_support['failed'] else 0
average_pe_oi_resistance_successful = sum(pe_oi_resistance['successful']) / len(pe_oi_resistance['successful']) if pe_oi_resistance['successful'] else 0
average_pe_oi_resistance_failed = sum(pe_oi_resistance['failed']) / len(pe_oi_resistance['failed']) if pe_oi_resistance['failed'] else 0

# Print results
print(f"Support Successful PCRs: {support_pcrs['successful']}")
print(f"Support Failed PCRs: {support_pcrs['failed']}")
print(f"Resistance Successful PCRs: {resistance_pcrs['successful']}")
print(f"Resistance Failed PCRs: {resistance_pcrs['failed']}")

print(f"Average intraday PCR for successful VWAP support: {average_support_successful_pcr:.2f}")
print(f"Average intraday PCR for failed VWAP support: {average_support_failed_pcr:.2f}")
print(f"Average intraday PCR for successful VWAP resistance: {average_resistance_successful_pcr:.2f}")
print(f"Average intraday PCR for failed VWAP resistance: {average_resistance_failed_pcr:.2f}")

print(f"Average ce_oi for successful VWAP support: {average_ce_oi_support_successful:.2f}")
print(f"Average ce_oi for failed VWAP support: {average_ce_oi_support_failed:.2f}")
print(f"Average pe_oi for successful VWAP resistance: {average_pe_oi_resistance_successful:.2f}")
print(f"Average pe_oi for failed VWAP resistance: {average_pe_oi_resistance_failed:.2f}")

# Count results
print(f"Support successful count: {support_counts['successful']}")
print(f"Support failed count: {support_counts['failed']}")
print(f"Resistance successful count: {resistance_counts['successful']}")
print(f"Resistance failed count: {resistance_counts['failed']}")

# Calculate success rates
support_success_rate = support_counts['successful'] / (support_counts['successful'] + support_counts['failed']) * 100 if support_counts['successful'] + support_counts['failed'] > 0 else 0
resistance_success_rate = resistance_counts['successful'] / (resistance_counts['successful'] + resistance_counts['failed']) * 100 if resistance_counts['successful'] + resistance_counts['failed'] > 0 else 0

print(f"Support success rate: {support_success_rate:.2f}%")
print(f"Resistance success rate: {resistance_success_rate:.2f}%")

```

```python
import pandas as pd

data = data_new
data.sort_values(by='datetime', inplace=True)
data['datetime'] = pd.to_datetime(data['datetime'])

# Use previous day's vah_ and val_
data['prev_day_vah'] = data['vah_'].shift(1)
data['prev_day_val'] = data['val_'].shift(1)
data.dropna(subset=['prev_day_vah', 'prev_day_val'], inplace=True)

trades = []

def calculate_trades_with_trailing_stoploss(df):
    entry_price = 0
    stop_loss = 0
    trade_type = None
    exit_reason = None
    vah_or_val_crossed = None
    vah_or_val_value = None
    target1 = target2 = target3 = None

    for index, row in df.iterrows():
        # Check for short selling entry
        if row['close'] > 1.0075 * row['prev_day_vah'] and trade_type is None:
            entry_price = row['close']
            stop_loss = entry_price * 1.015
            trade_type = 'short'
            entry_time = row['datetime']
            year = entry_time.year
            month = entry_time.month
            vah_or_val_crossed = 'prev_day_vah'
            vah_or_val_value = row['prev_day_vah']
            target1 = row['prev_day_vah']
            target2 = (row['prev_day_vah'] + row['prev_day_val']) / 2
            target3 = row['prev_day_val']

        # Check for long position entry
        elif row['close'] < 0.9925 * row['prev_day_val'] and trade_type is None:
            entry_price = row['close']
            stop_loss = entry_price * 0.985
            trade_type = 'long'
            entry_time = row['datetime']
            year = entry_time.year
            month = entry_time.month
            vah_or_val_crossed = 'prev_day_val'
            vah_or_val_value = row['prev_day_val']
            target1 = row['prev_day_val']
            target2 = (row['prev_day_vah'] + row['prev_day_val']) / 2
            target3 = row['prev_day_vah']

        if trade_type is not None:
            midpoint = (row['prev_day_vah'] + row['prev_day_val']) / 2

            if trade_type == 'short':
                if row['close'] <= target1:
                    stop_loss = entry_price
                if row['close'] <= target2:
                    stop_loss = row['prev_day_vah']
                if row['close'] <= target3:
                    stop_loss = midpoint

                if row['close'] >= stop_loss:
                    exit_price = row['close']
                    profit_loss = entry_price - exit_price
                    profit_loss_percentage = (profit_loss / entry_price) * 100
                    exit_reason = 'stop loss hit'
                    trades.append({
                        'entry_price': entry_price,
                        'exit_price': exit_price,
                        'profit_loss': profit_loss,
                        'trade_type': trade_type,
                        'entry_time': entry_time,
                        'exit_time': row['datetime'],
                        'year': year,
                        'month': month,
                        'profit_loss_percentage': profit_loss_percentage,
                        'vah_or_val_crossed': vah_or_val_crossed,
                        'vah_or_val_value': vah_or_val_value,
                        'exit_reason': exit_reason
                    })
                    entry_price = 0
                    stop_loss = 0
                    trade_type = None
                    exit_reason = None
                    vah_or_val_crossed = None
                    vah_or_val_value = None
                    target1 = target2 = target3 = None

            elif trade_type == 'long':
                if row['close'] >= target1:
                    stop_loss = entry_price
                if row['close'] >= target2:
                    stop_loss = row['prev_day_val']
                if row['close'] >= target3:
                    stop_loss = midpoint

                if row['close'] <= stop_loss:
                    exit_price = row['close']
                    profit_loss = exit_price - entry_price
                    profit_loss_percentage = (profit_loss / entry_price) * 100
                    exit_reason = 'stop loss hit'
                    trades.append({
                        'entry_price': entry_price,
                        'exit_price': exit_price,
                        'profit_loss': profit_loss,
                        'trade_type': trade_type,
                        'entry_time': entry_time,
                        'exit_time': row['datetime'],
                        'year': year,
                        'month': month,
                        'profit_loss_percentage': profit_loss_percentage,
                        'vah_or_val_crossed': vah_or_val_crossed,
                        'vah_or_val_value': vah_or_val_value,
                        'exit_reason': exit_reason
                    })
                    entry_price = 0
                    stop_loss = 0
                    trade_type = None
                    exit_reason = None
                    vah_or_val_crossed = None
                    vah_or_val_value = None
                    target1 = target2 = target3 = None

            # Exit at 15:15:00 if trade is still open
            if row['datetime'].time() >= pd.Timestamp('15:15:00').time() and trade_type is not None:
                exit_price = row['close']
                profit_loss = exit_price - entry_price if trade_type == 'long' else entry_price - exit_price
                profit_loss_percentage = (profit_loss / entry_price) * 100
                exit_reason = 'end of day'
                trades.append({
                    'entry_price': entry_price,
                    'exit_price': exit_price,
                    'profit_loss': profit_loss,
                    'trade_type': trade_type,
                    'entry_time': entry_time,
                    'exit_time': row['datetime'],
                    'year': year,
                    'month': month,
                    'profit_loss_percentage': profit_loss_percentage,
                    'vah_or_val_crossed': vah_or_val_crossed,
                    'vah_or_val_value': vah_or_val_value,
                    'exit_reason': exit_reason
                })
                entry_price = 0
                stop_loss = 0
                trade_type = None
                exit_reason = None
                vah_or_val_crossed = None
                vah_or_val_value = None
                target1 = target2 = target3 = None

calculate_trades_with_trailing_stoploss(data)
trades_df = pd.DataFrame(trades)

# Remove rows with 0 profit_loss_percentage
trades_df = trades_df[trades_df['profit_loss_percentage'] != 0]

yearly_profit_loss = trades_df.groupby('year')['profit_loss'].sum()
yearly_profit_loss_percentage = trades_df.groupby('year')['profit_loss_percentage'].sum()

monthly_profit_loss = trades_df.groupby(['year', 'month'])['profit_loss'].sum()
monthly_profit_loss_percentage = trades_df.groupby(['year', 'month'])['profit_loss_percentage'].sum()

total_profit_loss = trades_df['profit_loss'].sum()
total_profit_loss_percentage = trades_df['profit_loss_percentage'].sum()

print("\nYearly Profit/Loss Percentage:")
print(yearly_profit_loss_percentage)
print("\nMonthly Profit/Loss Percentage:")
print(monthly_profit_loss_percentage)
print("\nTotal Profit/Loss Percentage:")
print(total_profit_loss_percentage)
print("\nTrades DataFrame:")
#trades_df.head(60)
```

```python
import pandas as pd
import numpy as np

# Assuming `data_` is your initial DataFrame
data = data_oi.copy()  # Make a copy to avoid modifying original data directly
data['datetime'] = pd.to_datetime(data['datetime'])
data['date'] = data['datetime'].dt.date
data['year'] = data['datetime'].dt.year  # Extract year for grouping
data.sort_values(by='datetime', inplace=True)

# Function to calculate Point of Control (POC)
def calculate_poc(df):
    return np.sum(df['volume'] * (df['high'] + df['low'] + df['close']) / 3) / np.sum(df['volume'])

# Calculate POC for each day
daily_data = data.groupby('date').apply(calculate_poc).reset_index(name='poc')

# Identify days where the POC is trending higher or lower retrospectively
daily_data['poc_trending_higher'] = daily_data['poc'].shift(1) > daily_data['poc'].shift(2)
daily_data['poc_trending_lower'] = daily_data['poc'].shift(1) < daily_data['poc'].shift(2)

# Get dates where yesterday's POC trended higher or lower
trending_higher_dates = daily_data[daily_data['poc_trending_higher']]['date']
trending_lower_dates = daily_data[daily_data['poc_trending_lower']]['date']

results = []
near_ranges = [0.0125]

def get_previous_trading_day(date, all_dates):
    previous_dates = all_dates[all_dates < date]
    if not previous_dates.empty:
        return previous_dates.max()
    return None

unique_dates = pd.Series(data['date'].unique()).sort_values()

def process_day_with_trailing_stoploss(date, prev_date, strategy, local_extreme_fn, near_ranges):
    prev_day_data = data[data['date'] == prev_date]
    local_extreme = local_extreme_fn(prev_day_data['close'])
    daily_subset = data[data['date'] == date]

    entry_close = None
    stop_loss = None
    trailing_stop = None

    for near_range in near_ranges:
        within_range = daily_subset[
            (daily_subset['open'] <= local_extreme * (1 + near_range)) &
            (daily_subset['open'] >= local_extreme * (1 - near_range))
        ]

        if within_range.empty:
            continue

        entry_close = within_range['close'].iloc[-1]  # Last 5-min candle close within the range

        # Set initial stop-loss
        if strategy == 'Higher High POC':
            stop_loss = entry_close * 0.985  # 1.5% below entry for long position
        else:
            stop_loss = entry_close * 1.015  # 1.5% above entry for short position

        trailing_stop = stop_loss

        # Check Open Interest conditions
        if strategy == 'Higher High POC':
            ce_oi_condition = (
                daily_subset['ce_oi'].iloc[-1] > daily_subset['ce_oi'].shift(1).rolling(window=5).max()
            )
            ce_oi_condition = ce_oi_condition.all()  # Check if all conditions in the rolling window are True
            if not ce_oi_condition:
                continue  # Skip this iteration if ce_oi condition is not met
        else:
            pe_oi_condition = (
                daily_subset['pe_oi'].iloc[-1] > daily_subset['pe_oi'].shift(1).rolling(window=5).max()
            )
            pe_oi_condition = pe_oi_condition.all()  # Check if all conditions in the rolling window are True
            if not pe_oi_condition:
                continue  # Skip this iteration if pe_oi condition is not met

        # Check if stop-loss is breached or trailing stop is hit during the day
        for index, row in within_range.iterrows():
            if strategy == 'Higher High POC':
                if row['low'] <= trailing_stop:
                    movement_pct = (((trailing_stop - entry_close) / entry_close) * 100)
                    last_5min_close = trailing_stop
                    results.append({
                        'date': date,
                        'year': daily_subset.iloc[0]['datetime'].year,
                        'near_range': near_range,
                        'movement_pct': movement_pct,
                        'entry_close': entry_close,
                        'last_5min_close': last_5min_close,
                        'strategy': strategy
                    })
                    break  # Exit the loop if trailing stop is hit
                if row['low'] <= stop_loss:
                    movement_pct = (((stop_loss - entry_close) / entry_close) * 100)
                    last_5min_close = stop_loss
                    results.append({
                        'date': date,
                        'year': daily_subset.iloc[0]['datetime'].year,
                        'near_range': near_range,
                        'movement_pct': movement_pct,
                        'entry_close': entry_close,
                        'last_5min_close': last_5min_close,
                        'strategy': strategy
                    })
                    break  # Exit the loop if stop loss is hit
                trailing_stop = max(trailing_stop, row['close'] * 0.995)  # Update trailing stop
            else:
                if row['high'] >= trailing_stop:
                    movement_pct = -(((trailing_stop - entry_close) / entry_close) * 100)
                    last_5min_close = trailing_stop
                    results.append({
                        'date': date,
                        'year': daily_subset.iloc[0]['datetime'].year,
                        'near_range': near_range,
                        'movement_pct': movement_pct,
                        'entry_close': entry_close,
                        'last_5min_close': last_5min_close,
                        'strategy': strategy
                    })
                    break  # Exit the loop if trailing stop is hit
                if row['high'] >= stop_loss:
                    movement_pct = -(((entry_close - stop_loss) / entry_close) * 100)
                    last_5min_close = stop_loss
                    results.append({
                        'date': date,
                        'year': daily_subset.iloc[0]['datetime'].year,
                        'near_range': near_range,
                        'movement_pct': movement_pct,
                        'entry_close': entry_close,
                        'last_5min_close': last_5min_close,
                        'strategy': strategy
                    })
                    break  # Exit the loop if stop loss is hit
                trailing_stop = min(trailing_stop, row['close'] * 1.005)  # Update trailing stop

    if entry_close is not None:
        # If the trade is not closed due to hitting the stop-loss, exit at 15:15:00
        exit_time = pd.Timestamp(f'{date} 15:15:00')
        last_candle_before_exit = daily_subset[daily_subset['datetime'] <= exit_time].iloc[-1]
        last_5min_close = last_candle_before_exit['close']
        if strategy == 'Higher High POC':
            movement_pct = ((last_5min_close - entry_close) / entry_close) * 100
        else:
            movement_pct = -((last_5min_close - entry_close) / entry_close) * 100
        results.append({
            'date': date,
            'year': daily_subset.iloc[0]['datetime'].year,
            'near_range': near_range,
            'movement_pct': movement_pct,
            'entry_close': entry_close,
            'last_5min_close': last_5min_close,
            'strategy': strategy
        })

# Process higher and lower trending days with trailing stop-loss
for date in trending_higher_dates:
    prev_date = get_previous_trading_day(date, unique_dates)
    if prev_date:
        process_day_with_trailing_stoploss(date, prev_date, 'Higher High POC', min, near_ranges)

for date in trending_lower_dates:
    prev_date = get_previous_trading_day(date, unique_dates)
    if prev_date:
        process_day_with_trailing_stoploss(date, prev_date, 'Lower Low POC', max, near_ranges)

# Create DataFrame from results
results_df = pd.DataFrame(results)

# Calculate total percentage change including stop-loss hits
total_pct_change = results_df.groupby('near_range')['movement_pct'].sum()

# Print or use total_pct_change as needed
print("TOTAL PERCENTAGE CHANGE (Including stop-loss hits):")
print(total_pct_change)

yearly_profit_percentage = results_df.groupby('year')['movement_pct'].sum()

print("YEARLY PROFIT PERCENTAGE:")
print(yearly_profit_percentage)

bins = [-float('inf'), -2, -1, 0, 1, 1.5, 2, float('inf')]
labels = ['<-2%', '-2%--1%', '-1%-0%', '0%-1%', '1%-1.5%', '1.5%-2%', '>2%']
results_df['movement_range'] = pd.cut(results_df['movement_pct'], bins=bins, labels=labels)
range_counts = results_df.groupby(['near_range', 'movement_range']).size().unstack(fill_value=0)

print(" ")
print("TOTAL PERCENTAGE CHANGE:")
print(total_pct_change)
print(" ")
print("MOVEMENT PERCENTAGE RANGE COUNTS:")
print(range_counts)
```
